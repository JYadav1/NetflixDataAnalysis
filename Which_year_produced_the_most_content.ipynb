```python
# Import necessary libraries
import pandas as pd
import altair as alt
```

```python
# Load the dataset
df = pd.read_csv('netflix_titles.csv')
```

```python
#groupby to get the count of tv shows and movies
df1 = df.groupby(['type', 'release_year']).count().reset_index()
df1
```

```python
# Create a Grouped BAr Chat
alt.Chart(df1).mark_bar().encode(
    x=alt.X('type:O'),
    y='sum(show_id):Q',
    color='type:N',
    column='release_year:N'
)
```
