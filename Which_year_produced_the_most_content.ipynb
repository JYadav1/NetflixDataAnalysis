```python
# Import necessary libraries
import pandas as pd
import altair as alt
```

```python
# Load the dataset
df = pd.read_csv('netflix_titles.csv')
```

```python
#groupby to get the count of tv shows and movies
df1 = df.groupby(['type', 'release_year']).count().reset_index()
df1
```

```python
df1['type_year'] = list(zip(df1['type'], df1['release_year']))
df1
```

```python
# Create a Grouped BAr Chat
alt.Chart(df1).mark_bar().encode(
    x=alt.X('type:O'),
    y='sum(show_id):Q',
    color='type:N',
    column='release_year:N'
)
```

```python
df1.to_csv('netflix_titles.csv')
```

```python
df_type_year = df1#.groupby(['type', 'release_year']).count().reset_index()
df_movies = df_type_year[df_type_year['type'] == 'Movie']
df_shows = df_type_year[df_type_year['type'] == 'TV Show']
df_movies.set_index('release_year', inplace=True)
df_shows.set_index('release_year', inplace=True)
df_movies = df_movies.loc[2005:2020, :].reset_index()
df_shows = df_shows.loc[2005:2020, :].reset_index()
join = pd.concat([df_shows, df_movies],axis=0)
join
```

```python
# Create a Grouped Bar Chat
alt.Chart(join).mark_bar().encode(
    x=alt.X('type:O'),
    y='sum(show_id):Q',
    color='type:N',
    column='release_year:N'
)
```
