```python

import altair as alt
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('new_netflix.csv')
actors = pd.read_csv('actors.csv')
directors = pd.read_csv('directors.csv')
genres = pd.read_csv('Genres.csv')
```

```python
top10_actors = actors.head(10)
top10_directors = directors.head(10)

top10_actors
```

```python
alt.Chart(top10_actors).mark_bar().encode(
    x='Actor',
    y='Occurance'
)
```

```python
alt.Chart(top10_directors).mark_bar().encode(
    x='Director',
    y='Occurance'
)
```

```python
alt.Chart(genres).mark_bar().encode(
    x='Genre',
    y='Occurance'
)
```

```python
avg_sent = df.groupby('release_year')['Happy', 'Sad', 'Fear', 'Surprise', 'Angry'].mean()
avg_sent.index.name = 'Year'
avg_sent.to_csv('avg_sentiment.csv')

avg_sent = pd.read_csv('avg_sentiment.csv')
avg_sent = avg_sent[avg_sent['Year'] >= 1980]
avg_sent
```

```python
fear = alt.Chart(avg_sent).mark_line().encode(x='Year',y='Fear', color=alt.value("black"))
happy = alt.Chart(avg_sent).mark_line().encode(x='Year',y='Happy',color=alt.value("yellow"))
sad = alt.Chart(avg_sent).mark_line().encode(x='Year',y='Sad',color=alt.value("blue"))
surprise = alt.Chart(avg_sent).mark_line().encode(x='Year',y='Surprise',color=alt.value("green"))
angry = alt.Chart(avg_sent).mark_line().encode(x='Year',y='Angry',color=alt.value("red"))

sentiment = alt.layer(fear, happy, sad, angry, surprise)



sentiment
```

```python

alt.Chart(avg_sent).transform_fold(
    ['Fear', 'Happy', 'Sad', 'Surprise', 'Angry'],
).mark_line().encode(
    x='Year',
    y='value:Q',
    color='Sentiment:N'
)
```
